```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

np.random.seed(0)
random.seed(0)

data = {
    "ID": np.arange(1, 51),
    "Name": [f"Employee_{i}" for i in range(1, 51)],
    "Age": np.random.randint(20, 60, size=50),
    "Salary": np.random.randint(30000, 100000, size=50),
    "Department": np.random.choice(["HR", "Finance", "IT", "Marketing", "Sales"], size=50),
    "YearsExperience": np.random.randint(1, 40, size=50),
    "PerformanceScore": np.random.randint(1, 10, size=50),
    "Bonus": np.random.randint(1000, 10000, size=50),
    "JoinDate": [datetime.now() - timedelta(days=random.randint(100, 3650)) for _ in range(50)]
}
# print(data)
df = pd.DataFrame(data)
# Display the first 5 rows
print(df.head(20))

# df.plot(kind = 'line', x = 'Salary', y = 'YearsExperience')
# df.plot(kind='hist', y='Age', bins=20)
# df.plot(kind='hexbin', x='Age', y='Salary', gridsize=30) # Hexbin plot
# df.plot(kind='area', y='Salary')                # Area plot
df.plot(kind='pie', y='Salary', labels=df['Department']) # Pie chart
plt.show()

```
