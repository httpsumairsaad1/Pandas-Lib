```python
import pandas as pd

line = "=" * 60

df = pd.read_excel("E:/fake data/fake_data_with_duplicates.xlsx")
print(df, "\n", line, "duplicates", line)
print(df.duplicated())
duplicate_indices = df[df.duplicated()].index
# Print the index numbers
print("Indices Number: ", duplicate_indices)
print(line, "Drop duplicates", line)
df.drop_duplicates(inplace = True)
print(df)
```
